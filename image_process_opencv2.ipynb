```python
import cv2
import numpy as np
import sys
import os
import fnmatch
import matplotlib.pyplot as plt
```

```python
def resize(fname, width, height):
    image = cv2.imread(fname)
    org_height, org_width = image.shape[0:2]
    print("Height ", org_height)
    print("Width ", org_width)
    if org_width >= org_height:
        new_img = cv2.resize(image, (org_width,org_height))
    else:
        new_img = cv2.resize(image, (org_height,org_width))    
    # cv2.imshow("Original image", image)
    # cv2.waitKey(0)

    return fname, new_img
```

```python
filename,  new_image = resize("a.jpg", 1280,960)
plt.imshow(new_image)
```

```python
def sharpen(image):
    kernels = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
    new_img = cv2.filter2D(image,-1,kernels)
    # cv2.imshow("sharpened", new_img)
    # cv2.waitKey(0)
    return new_img 
```

```python
img_sharpen = sharpen(new_image)
plt.imshow(img_sharpen)
```

```python

def blur(image):
    kernels = [4]
    for idx, k in enumerate(kernels):
        image_bl = cv2.blur(image, ksize=(k,k))
        # cv2.imshow(str(k), image_bl)
        # cv2.waitKey(0)
    return image_bl
```

```python
img_blured = blur(img_sharpen)
plt.imshow(img_blured)
```
